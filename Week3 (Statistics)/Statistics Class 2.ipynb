# Degrees of Freedom

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)

# Generating population data
pop = np.random.normal(2,100,1000)

plt.hist(pop,bins=100)
plt.show()

print("Pop mean {} \n Pop variance {} \n Pop stddev {}".format(np.mean(pop),np.var(pop),np.std(pop)))
```

```python
# Generating sample data (with replacement)
#np.random.choice?
sample = np.random.choice(pop,100,replace=True)
```

```python
plt.hist(sample,bins=30)
plt.show()
```

```python
# Computing the sample mean and variance
n = len(sample)
print("Sample mean {} \n Sample variance {} \n Sample StdDev {}".format(np.mean(sample),np.var(sample)*(n/(n-1)),np.sqrt(np.var(sample)*(n/(n-1)))))
```

```python
# Generating samples and their moments
sample_mean = []
sample_var  = []
sample_var_biased = []
n = 100 # length of sample
for i in range(10000):
    sample = np.random.choice(pop,n,replace=True)
    sample_mean.append(np.mean(sample))
    sample_var.append(np.var(sample)*(n/(n-1))) # unbiased estimate
    sample_var_biased.append(np.var(sample)) # biased estimate of the sample variance    
```

```python
# Plotting sample means
plt.hist(sample_mean,bins=100)
plt.vlines(np.mean(pop),0,350,label="Population Mean",color='y')
plt.vlines(np.mean(sample_mean),0,350,label="Grand Sample Mean",color='r') # Mean of all sample means
plt.legend();plt.show()

print(np.mean(sample_mean),np.mean(pop))
```

```python
fig,ax = plt.subplots(1,2,figsize=[20,8])

# Plotting sample variances
ax[0].hist(sample_var,bins=100)
ax[0].vlines(np.var(pop),0,350,label="Population Variance",color='y')
ax[0].vlines(np.mean(sample_var),0,350,label="Variance Mean (Unbiased)",color='r') # Mean of all sample means
ax[0].vlines(np.mean(sample_var_biased),0,350,label="Variance Mean (Biased)",color='k') # Mean of all sample means
ax[0].legend()

ax[1].vlines(np.var(pop),0,350,label="Population Variance",color='y')
ax[1].vlines(np.mean(sample_var),0,350,label="Variance Mean (Unbiased)",color='r') # Mean of all sample means
ax[1].vlines(np.mean(sample_var_biased),0,350,label="Variance Mean (Biased)",color='k') # Mean of all sample means
ax[1].legend()


plt.show()

print(np.var(pop),np.mean(sample_var))
```

- Repeated sampling (bootstrapping) gives us a very close estimate of the population mean and variance
